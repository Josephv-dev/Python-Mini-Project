```python
import ipywidgets as widgets
from IPython.display import display
import time
import datetime
import playsound  # Or winsound on Windows pip install plysound

def set_alarm(alarm_time_str): #it takes the alarm time as a string
    try:
        # Attempt to parse the time string directly (e.g., "7:30 AM", "19:30")
        alarm_time = datetime.datetime.strptime(alarm_time_str, '%I:%M %p')  # 12-hour format
    except ValueError:
        try:
            alarm_time = datetime.datetime.strptime(alarm_time_str, '%H:%M')  # 24-hour format
        except ValueError:
            print("Invalid time format. Please use 'HH:MM AM/PM' or 'HH:MM'.")
            return  # Exit the function if the format is invalid

    now = datetime.datetime.now()
    alarm_time = alarm_time.replace(year=now.year, month=now.month, day=now.day)  # Set date

    if alarm_time < now:  # If alarm time is in the past, set for next day
        alarm_time = alarm_time + datetime.timedelta(days=1)

    print(f"Alarm set for {alarm_time}")

    while datetime.datetime.now() < alarm_time:
        time.sleep(1)  # Check every second

    print("Wake up!")  # Simulate alarm going off
    try:
        playsound.playsound(r"C:\Users\DEL\Music\music\Creepin'.mp3")  # Replace with your own sound file, make sure to specify the correct path
    except Exception as e:
        print(f"Error playing sound: {e}")


# Create widgets
time_input = widgets.Text(placeholder="Enter alarm time (e.g., 7:30 AM or 19:30)", description="Time:")
set_button = widgets.Button(description="Set Alarm")
message_output = widgets.Output()  # Use Output widget for better message display


def on_set_clicked(b):
    alarm_time_str = time_input.value
    with message_output:  # Use the Output widget as a context manager
        message_output.clear_output() # Clear previous messages
        set_alarm(alarm_time_str)

set_button.on_click(on_set_clicked)

# Display widgets
display(time_input)
display(set_button)
display(message_output)
```
